# Code for train/test split to use for every models.

This ensures data consistency for all models.

```python
import pandas as pd

# Adjust the path depending on your file structure
df = pd.read_csv("../processed_online_shoppers_data.csv")

X = df.drop("Revenue", axis=1)
y = df["Revenue"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```
